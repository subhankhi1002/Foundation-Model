```python
def calculate_s(user_id, parent_asin, df_common, df_user_parent_asins):
    s_value = 0
    
    if user_id in df_user_parent_asins['user_id'].values:
        parent_asins = df_user_parent_asins[df_user_parent_asins['user_id'] == user_id]['parent_asin'].values[:1]
        
        for p in parent_asins:
            
            pe = df_common.loc[p, p]
            pej = df_common.loc[parent_asin, parent_asin]
            pe_ej = df_common.loc[parent_asin, p]
            
            numerator = 2 * pe_ej
            denominator = pe + pej
            
           
            s_value += numerator / denominator if denominator != 0 else 0
            
            
                
    return s_value




j = 0

user_id='new_user'
for parent_asin in df_common_main.columns:
    s_main_category = calculate_s(user_id, parent_asin, df_common_main, df_user_parent_asins)
#     s_store = calculate_s(user_id, parent_asin, df_common_store, df_user_parent_asins)
    s_categories = calculate_s(user_id, parent_asin, df_common_cat, df_user_parent_asins)

    df_s_main_category.at[user_id, parent_asin] = s_main_category
#     df_s_store.at[user_id, parent_asin] = s_store
    df_s_categories.at[user_id, parent_asin] = s_categories
#     print(j)
#     j+=1
```

```python
import numpy as np
from sklearn.decomposition import NMF

# Convert DataFrame to numpy array
R = df_s_categories.values

# Rank of the approximation
k = 50

# Initialize NMF model with non-negativity constraints
model = NMF(n_components=k, init='random', random_state=0)
U_cat = model.fit_transform(R)
V_cat = model.components_
```

```python
import numpy as np
from sklearn.decomposition import NMF

# Convert DataFrame to numpy array
R_main = df_s_main_category.values

# Rank of the approximation
k = 50

# Initialize NMF model with non-negativity constraints
model = NMF(n_components=k, init='random', random_state=0)
U_main = model.fit_transform(R_main)
V_main = model.components_
```

```python
import numpy as np
R_hat = np.dot(U_cat, V_cat)

# Compute the Frobenius norm of the residual matrix
frobenius_norm = np.linalg.norm(R_main - R_hat, 'fro')

# Calculate the singular values (sigma)
sigma_cat = np.sqrt(frobenius_norm)
```

```python
import numpy as np
R_hat = np.dot(U_main, V_main)

# Compute the Frobenius norm of the residual matrix
frobenius_norm = np.linalg.norm(R_main - R_hat, 'fro')

# Calculate the singular values (sigma)
sigma_main = np.sqrt(frobenius_norm)
```

```python
def compute_similarity(U, V, U_main, V_main, theta,theta_main, i):
    similarities = []
    for j in range(V.shape[1]):  # Iterate over columns of V
        dot_product = np.dot(U[i], V[:, j])  # Use all columns of V for the dot product
#         dot_product_store = np.dot(U_store[i], V_store[:, j])
        dot_product_main = np.dot(U_main[i], V_main[:, j])
        similarity = np.sum(theta * dot_product) +theta_main*dot_product_main
        similarities.append((j, similarity))
      
    
    # Sort similarities based on the second element (similarity value)
    similarities.sort(key=lambda x: x[1], reverse=True)
    
    return similarities

# Extract the row corresponding to the given user_id from df_s_categories
user_id = 'new_user' # Replace with your user_id
ui_index = df_s_categories.index.get_loc(user_id)

theta = sigma_cat

theta_main=sigma_main

# Compute similarities for ui_index
similarities = compute_similarity(U_cat, V_cat,U_main, V_main, theta,theta_main, ui_index)
```

```python
top_10_products = []
for j, similarity in similarities[:20]:
    product_id = df_s_categories.columns[j]
    title = df.loc[df['parent_asin'] == product_id, 'title'].iloc[0]
    top_10_products.append((product_id, title, similarity))

# Create a DataFrame from top 10 products
result_df = pd.DataFrame(top_10_products, columns=['product_id', 'title', 'similarity'])

print(result_df)
```
