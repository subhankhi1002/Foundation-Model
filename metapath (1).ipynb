```python
def merge_categories(series):
    merged_set = set()
    for s in series:
        merged_set.add(s)
    return merged_set

# Group by 'parent_asin' and aggregate 'main_category' for each parent_asin
category_groups = df.groupby('parent_asin')['main_category'].agg(merge_categories).reset_index()

# Create a dictionary to store category sets for each parent_asin
category_dict = category_groups.set_index('parent_asin').to_dict()['main_category']

# Initialize a dictionary to store the count of common categories
common_main_category = {}

# Iterate through combinations of parent_asin to find common categories
i = 0
for idx1, asin1 in enumerate(df['parent_asin'].unique()):
    common_main_category[asin1] = {}
    for idx2, asin2 in enumerate(df['parent_asin'].unique()):
        if idx1 == idx2:
            # Count the number of unique categories for asin1
            common_main_category[asin1][asin2] = len(category_dict[asin1])
        else:
            # Count the number of common categories between asin1 and asin2
            common_count = len(category_dict[asin1].intersection(category_dict[asin2]))
            common_main_category[asin1][asin2] = common_count

    print("row: ", i)
    i += 1

# Convert dictionary to DataFrame
df_common_main_category = pd.DataFrame(common_main_category)
```

```python
def calculate_s(user_id, parent_asin, df_common, df_user_parent_asins):
    s_value = 0
    
    if user_id in df_user_parent_asins['user_id'].values:
        parent_asins = df_user_parent_asins[df_user_parent_asins['user_id'] == user_id]['parent_asin'].values[:1]
        
        for p in parent_asins:
            pe = df_common.loc[p, p]
            pej = df_common.loc[parent_asin, parent_asin]
            pe_ej = df_common.loc[parent_asin, p]
            
            numerator = 2 * pe_ej
            denominator = pe + pej
            
            s_value += numerator / denominator if denominator != 0 else 0
                
    return s_value
```

```python
unique_user_ids = df['user_id'].unique()
unique_parent_asins = df['parent_asin'].unique()
df_s_main_category = pd.DataFrame(index=unique_user_ids, columns=unique_parent_asins)
df_s_store = pd.DataFrame(index=unique_user_ids, columns=unique_parent_asins)
df_s_categories = pd.DataFrame(index=unique_user_ids, columns=unique_parent_asins)

# Create df_user_parent_asins DataFrame
df_user_parent_asins = df[['user_id', 'parent_asin']].drop_duplicates()
```

```python
for i, user_id in enumerate(unique_user_ids):
    for parent_asin in unique_parent_asins:
        s_main_category = calculate_s(user_id, parent_asin, df_common_main_category, df_user_parent_asins)
        s_store = calculate_s(user_id, parent_asin, df_common_store, df_user_parent_asins)
        s_categories = calculate_s(user_id, parent_asin, df_common_category, df_user_parent_asins)
        
        df_s_main_category.at[user_id, parent_asin] = s_main_category
        df_s_store.at[user_id, parent_asin] = s_store
        df_s_categories.at[user_id, parent_asin] = s_categories
    print(i)
```

```python
def compute_similarity(U, V, U_store, V_store, theta, theta_store, i):
    similarities = []
    for j in range(V.shape[1]):  # Iterate over columns of V
        dot_product = np.dot(U[i], V[:, j])  # Use all columns of V for the dot product
        dot_product_store = np.dot(U_store[i], V_store[:, j])
        similarity = np.sum(theta * dot_product) + theta_store * dot_product_store
        similarities.append((j, similarity))
    
    # Sort similarities based on the second element (similarity value)
    similarities.sort(key=lambda x: x[1], reverse=True)
    
    return similarities

# Extract the row corresponding to the given user_id from df_s_categories
user_id = 'AF4VP45PTKQ7BO4F43GHIN6K4L6Q'  # Replace with your user_id
ui_index = df_s_categories.index.get_loc(user_id)

theta = np.array([0.5, 0.6])
theta_store = 0.7

# Compute similarities for ui_index
similarities = compute_similarity(U, V, U_store, V_store, theta, theta_store, ui_index)

# Retrieve top 10 products with their titles
top_10_products = []
for j, similarity in similarities[:0]:
    product_id = df_s_categories.columns[j]
    title = df.loc[df['parent_asin'] == product_id, 'title'].iloc[0]
    top_10_products.append((product_id, title, similarity))

# Create a DataFrame from top 10 products
result_df = pd.DataFrame(top_10_products, columns=['product_id', 'title', 'similarity'])

print(result_df)
```
